# Sales per States

```python
state=df['State'].unique()
sales_state=np.array([df['Sales'][df['State']==i].sum()/df['Sales'].sum() for i in state])
plt.figure(figsize=(30,10))
sns.barplot(x=state,y=sales_state*100)
plt.xticks(rotation=90)
plt.title('Sales per State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.show()
```

# Sales per Region

```python
region=df['Region'].unique()
sales_region=np.array([df['Sales'][df['Region']==i].sum()/df['Sales'].sum() for i in region])
plt.figure(figsize=(30,15))
sns.barplot(x=region,y=sales_region*100)
plt.xticks(rotation=90)
plt.title('Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

Inference:-Sales in West region are more.
    Sales in South need Improvisation

# Profit in Region

```python
region=df['Region'].unique()
profit_region=np.array([df['Profit'][df['Region']==i].sum()/df['Profit'].sum() for i in region])
plt.figure(figsize=(30,15))
sns.barplot(x=region,y=profit_region*100)
plt.xticks(rotation=90)
plt.title('Sales per Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.show()
```

Inference:-Profit in West region are more.
    Profit in Central & South need Improvisation

# Profit in State

```python
state=df['State'].unique()
profit_state=np.array([df['Profit'][df['State']==i].sum()/df['Profit'].sum() for i in state])
plt.figure(figsize=(30,10))
sns.barplot(x=state,y=profit_state*100)
plt.xticks(rotation=90)
plt.title('Sales per State')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.show()
```
